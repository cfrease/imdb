# How many movies came out by year? 

```python
def movie_count(data):
    Total = df['Year'].value_counts()
    Percentage = df['Year'].value_counts(normalize=True)
    
    return pd.concat([Total,Percentage], axis=1, keys = ['Total', 'Percent'])
movie_count(df)
```

# Which movies made less than the yearly average?

```python
y_rev = df.groupby(['Year'])['Revenue (Millions)'].mean().to_dict()
y_rev
```

```python
y_rev['2006']
```

```python
def find_rev(revenue, year):
    return revenue<y_rev[year]
    
new_df = df[df.apply(lambda x : find_rev(x['Revenue (Millions)'],x['Year']),axis=1)]

new_df
```
